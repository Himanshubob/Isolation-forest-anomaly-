```python
df=pd.read_csv('Healthcare-Diabetes.csv')
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf=IsolationForest(contamination='auto')
```

```python
clf.fit(df)
```

```python
predictions=clf.predict(df)
```

```python
predictions
```
